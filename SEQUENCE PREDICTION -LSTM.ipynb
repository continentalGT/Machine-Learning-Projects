# SEQUENCE PREDICTION THROUGH LSTM

```python
# 1 2 3 4 5 ->6
```

# 1. Importing Libraries

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

## 2.Generating data and target

```python
data  = [[[(i+j)/100] for i in range(5)] for j in range(100)]
data[:][0]
```

```python
target = [x/100 for x in range(5,105)]
target[:6]
```

Converting list to Array

```python
#array
data=np.array(data)
target=np.array(target)
type(data)
```

```python
data.shape,target.shape
```

TRAIN TEST SPLIT

```python
x_train,x_test,y_train,y_test = train_test_split(data,target,test_size=0.2,random_state=98)
x_train.shape,y_test.shape
```

# 3.model building

```python
#LSTM
model = Sequential()
model.add(LSTM((1),batch_input_shape=(None,5,1),return_sequences=False))
model.compile(loss='mean_absolute_error',optimizer='adam',metrics=['accuracy'])
model.summary()
```

In the context of LSTM (Long Short-Term Memory) models, the batch input shape parameter refers to the shape of the input data that is fed into the LSTM network during training or inference. The batch input shape is typically a three-dimensional array with the following dimensions: (batch_size, timesteps, features).

batch_size: This represents the number of sequences or samples in each batch. It indicates how many sequences are processed together in parallel during training. Each sequence in the batch can be considered as an independent input to the LSTM network.

timesteps: This refers to the number of time steps or sequence length in each input sequence. It represents how many time steps or elements are there in each sequence.

features: This represents the number of features or input dimensions at each time step. It indicates the number of variables or attributes you have for each element in the sequence.

For example, if you have a batch of 32 sequences, each with 10 time steps and 3 features, the batch input shape parameter for the LSTM network would be (32, 10, 3).

```python
history = model.fit(x_train,y_train,epochs=500,validation_data=(x_test,y_test))
```

```python
prediction=model.predict(x_test)
prediction
```

```python
from sklearn.metrics import *
```

```python
mean_absolute_error(y_test,prediction)
```

```python
plt.scatter(range(20),y_test,color='red')
plt.scatter(range(20),prediction,color='Blue');
```

```python
plt.plot(history.history['loss'])
```

```python
def guess():
    n = int(input())
    seq=np.array([[n],[n+1],[n+2],[n+3],[n+4]]).reshape(1,5,1)
    seq=seq/100
    model.predict(seq)
    print(seq.shape)
    return model.predict(seq)*100
```
